```python
import os
import subprocess
import datetime

base_detectable = 'toses/DETECTABLE'
base_no_detectable = 'toses/NO-DETECTABLE'

sub_dir_detectable = []
sub_dir_no_detectable = []

class Archivo:
    def __init__(self, file, path):
        self.file = file
        self.path = path

with os.scandir(base_detectable) as ficheros:            
    sub_dir_detectable = [fichero.name for fichero in ficheros if fichero.is_dir()]
    
with os.scandir(base_no_detectable) as ficheros:      
    sub_dir_no_detectable = [fichero.name for fichero in ficheros if fichero.is_dir()]

audio_files_detectable = []
audio_files_no_detectable = []

for x in sub_dir_detectable:
    with os.scandir(base_detectable+'/'+x) as sub_dir:
        for file in sub_dir:
            if file.is_file():
                f = Archivo(file.name, base_detectable+'/'+x+'/')   
                audio_files_detectable.append(f)

for x in sub_dir_no_detectable:
    with os.scandir(base_no_detectable+'/'+x) as sub_dir:
        for file in sub_dir:
            if file.is_file():
                f = Archivo(file.name, base_no_detectable+'/'+x+'/')                             
                audio_files_no_detectable.append(f)
                
#for det in audio_files_detectable:
#    print(det.file)
#    print(det.path)
#    print(det.file.replace('.ogg','.wav'))
    
#for det in audio_files_no_detectable:
#    print(det.file)
#    print(det.path)
#    print(det.file.replace('.ogg','.wav'))

print(datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"))



```

```python
#DETECTABLE
ok = 0
no_ok = 0
print('Start conversion from .ogg to .wav ' + datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
for det in audio_files_detectable:
    
    file_name = det.file
    file_path = det.path
    new_file_name = det.file.replace('.ogg','.wav')
    file_origin = file_path+file_name
    file_dest = base_detectable+'/'+new_file_name
    
    process = subprocess.run(['c:\\dev\\ffmpeg\\bin\\ffmpeg', '-i', file_origin, file_dest])
    #print(file_origin+'    '+file_dest)
    if process.returncode != 0:
        no_ok = no_ok + 1
    else:
        ok = ok + 1
    
print('Finish conversion from .ogg to .wav ' + datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
print('Ok=' + str(ok))
print('No Ok=' + str(no_ok))
```

```python
#NO DETECTABLE
ok = 0
no_ok = 0
print('Start conversion from .ogg to .wav ' + datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
for det in audio_files_no_detectable:
    
    file_name = det.file
    file_path = det.path
    new_file_name = det.file.replace('.ogg','.wav')
    file_origin = file_path+file_name
    file_dest = base_no_detectable+'/'+new_file_name
    
    process = subprocess.run(['c:\\dev\\ffmpeg\\bin\\ffmpeg', '-i', file_origin, file_dest])
    #print(file_origin+'    '+file_dest)
    
    if process.returncode != 0:
        no_ok = no_ok + 1
    else:
        ok = ok + 1
    
print('Finish conversion from .ogg to .wav ' + datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
print('Ok=' + str(ok))
print('No Ok=' + str(no_ok))
```

```python
# Genero listado de archivos DETECTABLES a procesar para generacin de imgenes
import librosa
import librosa.display
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas

print('Start image generation ' + datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
i = 0
path_destination = 'toses-images/DETECTABLE/'

with os.scandir(base_detectable) as sub_dir:
        for file in sub_dir:
            if file.is_file():
                i = i + 1
                
                filename = file.name.replace('.wav','')
                file = filename + '.wav'
                samples, sample_rate = librosa.load(base_detectable + '/' + file)

                fig = plt.figure(figsize=[4,4]) #4,4 original value
                plt.axis('off')

                S = librosa.feature.melspectrogram(y=samples, sr=sample_rate)

                canvas = FigureCanvas(fig)
                p = librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
                fig.savefig(path_destination + filename + '.png', bbox_inches='tight',pad_inches = 0)
                plt.clf()
plt.close()                
print(i)
print('Finish image generation ' + datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
```
