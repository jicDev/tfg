```python
import librosa
import librosa.display
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas

filename = 'WhatsApp Audio 2022-06-07 at 10.33.36 AM'
file = filename + '.wav'
samples, sample_rate = librosa.load('audio-prueba-novia-juan/' + file)

fig = plt.figure(figsize=[1,1]) #4,4 original value
plt.axis('off')

S = librosa.feature.melspectrogram(y=samples, sr=sample_rate)

canvas = FigureCanvas(fig)
p = librosa.display.specshow(librosa.power_to_db(S, ref=np.max))
fig.savefig('audio-prueba-novia-juan/' + filename + '.png', bbox_inches='tight',pad_inches = 0)
```
